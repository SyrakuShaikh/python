```python
# pure Python
def py_int(a, b, f, N=2000):
    dx = (b-a)/N
    s = 0.0
    for i in range(N):
        s += f(a+i*dx)
    return s * dx

from math import pi, sin

def sin2(x):
    return sin(x) ** 2

def py_main():
    a, b = 0.0, 2.0 * pi
    return py_int(a, b, sin2, N=400000)
```

```python
import cProfile
cProfile.run('py_main()', sort='time')
```
